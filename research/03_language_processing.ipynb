```python
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
nlp = spacy.load("en_core_web_lg")
```

```python

text_columns = ['URL', 'Title', 'TitleAndDate', 'Texts']

def identify_entities(text):
    """
    Process the text with spaCy and extract entities.
    """
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

# Initialize an empty set to store unique entities
all_unique_entities = set()

# Iterate through each text-containing column and extract entities
for column in text_columns:
    if column in df.columns:
        # Extract entities for each row in the column and add to the set
        df[column].astype(str).apply(lambda x: all_unique_entities.update(identify_entities(x)))

# Convert the set back to a list if you need list operations later
all_unique_entities_list = list(all_unique_entities)

# Display the unique entities
print(all_unique_entities_list)
```

```python
# Function to identify potential product mentions
def identify_products(text):
    doc = nlp(text)
    # Extract entities potentially related to products
    # You might choose to focus on 'PRODUCT' labels, or include 'ORG' for branded products
    products = [ent.text for ent in doc.ents if ent.label_ in ['PRODUCT', 'ORG']]
    return products

# Initialize a set to store unique product mentions
unique_products = set()

# Iterate through each text-containing column and extract product mentions
for column in text_columns:
    if column in df.columns:
        # Update the set of unique products with mentions found in this column
        df[column].astype(str).apply(lambda x: unique_products.update(identify_products(x)))

# Convert the set of unique products to a sorted list for display
unique_products_list = sorted(list(unique_products))

# Display the list of unique product mentions
print(unique_products_list)
```

```python
keywords = ["launch", "introduce", "brand", "model", "release"]
from spacy.matcher import Matcher

# Initialize the Matcher with the shared vocab
matcher = Matcher(nlp.vocab)

# Add match patterns for keywords
for keyword in keywords:
    pattern = [{"LOWER": keyword}]
    matcher.add(f"KEYWORD_{keyword.upper()}", [pattern])

def extract_potential_products(text):
    doc = nlp(text)
    # Find matches
    matches = matcher(doc)
    potential_products = []
    for match_id, start, end in matches:
        # Extract the sentence containing the matched keyword
        sentence = doc[start].sent
        # Extract noun chunks from the sentence as potential product mentions
        for chunk in sentence.noun_chunks:
            potential_products.append(chunk.text)
    return list(set(potential_products))  # Remove duplicates

# Example usage with a single text snippet
text = "Apple is set to launch the iPhone 13 this fall, introducing a new model with improved features."
potential_products = extract_potential_products(text)
print(potential_products)
```
